```python
r,g,b = cat_arr[500,300] # tuple unpacking
print("Red level =", r, "Green level =", g, "Blue level =", b)
```

```python
# slice by color index
red_level = cat_arr[500,300,0] 
green_level = cat_arr[500,300,1]
blue_level = cat_arr[500,300,2]
print(red_level,green_level,blue_level)
```

```python
cat_only_green = cat_arr.copy() 
```

```python
cat_only_green[:,:,(0,2)] = 0
```

```python
pyplot.imshow(cat_only_green)
```

```python
cat_only_blue = cat_arr.copy() 
```

```python
cat_only_blue[:,:,(0,1)] = 0
```

```python
pyplot.imshow(cat_only_blue)
```

```python
# in order to concatenate images using numpy, use: concatenate or stack
cat_gallery= np.hstack((cat_only_red,cat_only_green,cat_only_blue))
```

```python
pyplot.imshow(cat_gallery)
```

```python
cat_gallery2= np.concatenate((cat_only_red,cat_only_green,cat_only_blue),axis=1)
```

```python
pyplot.imshow(cat_gallery2)
```
